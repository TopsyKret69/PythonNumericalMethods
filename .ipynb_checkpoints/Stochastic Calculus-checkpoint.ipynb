## Brownian Motion/Wiener Process

The stochastic process $W_t$ is called a Wiener process/Brownian motion if it satisfies the following conditions

* $W_0$ = 0
* $W_t-W_s$ independent of $W_s$ $\forall s<t$
* $W_t$ is normally distributed with mean $0$ and variance $t$
* $W_t$ has continuous paths



## Simulation of Wiener Process

* $\underline{z} = [z_1,z_2,\dots, z_n]$ random vector, containing independent, std normal variables
* by definition $\tilde{W_t} = 0$
* it has normal increments with mean $0$ and $t$ variance $\Rightarrow$ $\tilde{W}_{t+\Delta t} = \tilde{W}_{t} +z_n \sqrt{\Delta t}$
* therefore $\tilde{W}_{t} = \sum_{i=1}^n \sqrt{\Delta t} z_i$



```python
import numpy as np
import numpy.random as rnd
import matplotlib.pyplot as plt
```

```python
rnd.seed(1729)

#numer or random numbers
N = 100
#number of Wiener process simulations
M = 20
#get the random numbers
z = rnd.standard_normal([M,N-1])

w_t = np.zeros([M,N])
w2_t = np.zeros([M,N])
for i in np.arange(1,N):
    w_t[:,i] = w_t[:,i-1] + np.sqrt(1/N) * z[:,i-1]
    w2_t[:,i] = w_t[:,i]**2

#let's display the paths
for j in np.arange(0,M-1):
    plt.plot(w_t[j,:])
    
plt.ylabel("trajectory")
plt.show()
for j in np.arange(0,M-1):
    plt.plot(w2_t[j,:])
plt.ylabel("w2 trajectory")    
plt.show()
```

## Integration with respect to $W_t$

Let's consider the following integral
$$ \int_0^1 a(t) dW_t $$

* This is called stochastic integral with respect to Wiener process.
* The integral is dependent on $a(t)$ of course 
##### Theorem:
* Let $a(t)$ be an adapted stochastic process such that $\mathbb{E}\left(\int_0^T a(t)dt\right)<\infty$ and $T < \infty$
* Then $\int_0^T a(t) dW_t = lim_{n\to \inf} \sum_0^n a(t_i)\left( W_{t_{i+1}} - W_{t}\right)$ is well defined integral with the following properties

1) continuous

2) linear 

3) martingale

4) etc


#### Example:
Let's calculate numerically
$$\int_0^1 W_tdW_t$$

```python


#Let's integrate w_t[0,:] with respect to w_t[0,:]
W = np.zeros([M,N])
for k in np.arange(1,N-1):
    W[:,k] = W[:,k-1] + w_t[:,k-1] * (w_t[:,k] - w_t[:,k-1])

for j in range(0,M-1):
    plt.plot(W[j,:])
plt.show()
    
W2 = np.zeros([M,N])
for k in np.arange(0,M-1):
    W2[k,:] = w_t[k,:] * w_t[k,:]    


plt.plot(W2[0,:])    
plt.show()
```

### Ito's Formula

If $f(t,x) \in$ $C^{1,2}$, and $W_t$ is the Wiener-process, then
$$
f(T,W_T) - f(0,W_0) = \int_0^T f_x(t,W_t)dW_t + \frac{1}{2}\int_0^Tf_{xx}(t,W_t)d[W]_t + \int_0^Tf_t(t,W_t)dt
$$

Therefore, assuming $f(t,x) = x^2$ then

$$
W_T^2 - W_0^2 = \int_0^T 2W_t dW_t + \frac{1}{2}\int_0^T2dt
$$
Result:

$$
\frac{1}{2}\left(W_T^2 - T\right)  = \int_0^T W_t dW_t 
$$
