```python
#using pandas 

data = pd.read_csv("MOL.BD.csv")
print(data.head())
print("Start date " + data["Date"][0])
print(data["Date"].describe())
#print(data["Close"])

import matplotlib.pyplot as plt
#visualizing the stock price
plt.plot(data["Close"])
plt.ylabel("Stock Closing Prices")
plt.show()

```

```python
#Let's see the 1 day log differences of stock prices
import numpy as np

stockPrice = np.array(data["Close"])

logReturns = np.log(stockPrice[1:]) - np.log(stockPrice[:-1])
print(logReturns[1:4])
print(np.log(data["Close"][3]/data["Close"][2]))

plt.plot(logReturns)
plt.show()

plt.hist(logReturns[1:-1],bins=20)
plt.show()

mean = np.mean(logReturns)
stdDev = np.std(logReturns)

normalizedLogReturns = (logReturns - mean)/stdDev

plt.plot(normalizedLogReturns)
plt.show()

plt.hist(normalizedLogReturns,bins=20)

print(np.mean(logReturns))
print(np.std(logReturns))
```
